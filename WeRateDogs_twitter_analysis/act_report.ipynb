## Insights from WeRateDogs Data


WeRateDogs is one of the most popular Twitter accounts with over 9 million followers at the time of writing this report. This account is devoted to rating dogs. Here is the twitter profile page of WeRateDogs:




You can probably see the appeal of the account especially if you're a dog person. Even if you aren't particularly fond of dogs, you have to admit that a lot of the reviewed  dogs are pretty cute. Tweets made by the account follow a similar format with a description of the dog, a score and some form of media (usually a picture) depicting the dog.

In this article, I'll be taking a look at the archived tweets graciously provided by WeRateDogs beginning from the account's creation up to around August of 2017. The goal is to illustrate the process of gaining insight into the data after a thorough wrangling process. In addition to the archived tweet data, I was also provided with the prediction results of a machine learning algorithm trained on pictures from a sample of the archived tweets which I used to supplement the archived tweet data.

I've already performed the tedious but necessary task of wrangling the data so I'll jump right into sharing some of the insights I've gained from the data which I've grouped into two categories: insights about the tweets themselves and general insights about the WeRateDogs account. 



### Tweet insights
#### Most common dog breeds
One of the first things I was interested in knowing is: what are the most common dog breeds that appear in the tweets made by WeRateDogs? Since the cleaned dataset already contains the correct predictions made by the algorithm, I've been spared having to trawl through the over 2000 pictures and identify the breed of each dog. I should mention that not every tweet in the dataset has a corresponding dog breed. This is to be expected given the nature of machine learning algorithms and the process of training them. This isn't something that I need to be too concerned about as the total amount of correct predictions made by the algorithm is sufficient for the insight gained here to be meaningful.

            

The visualization above shows the dog breed that appears most often in WeRateDogs tweets even compared to the rest of the most common dog breeds is the Golden Retriever. In fact, it seems that the Golden Retriever is far and away the most common dog breed. Given the WeRateDogs account's huge following, this could be an indication of the overall popularity of not only the breeds shown in the chart but the other breeds in the sample as well. This is, of course, assuming that the people who follow WeRateDogs is representative of dog owners as a whole.

#### Most common dog type
One of the unique aspects of WeRateDogs is the lingo that has developed around the account. One of the most common and fun bits of this lingo is how people refer to dogs. Dogs are called 'pupper', 'puppo', or 'doggo' based on how old the dog is with 'pupper' being used for puppies, 'doggo' for adult dogs, and 'puppo' for dogs in a stage of development between puppy and adult. There is also the word 'floofer' used to describe dogs of any age with a particularly fluffy coat. A visualizaton was created to see which of these words appears the most in the archived tweets.

  

From the visualization above, It appears that the most commonly used term is 'pupper' indicating that the majority of the dogs shown in WeRateDogs are puppies. 

#### Most common scores

The scoring style of WeRateDogs made me draw insights from the the most common scores given to the dogs shown in each tweet.
    

The result shows that the most common score by far is 12/10 followed by 10/10 and 11/10. It's interesting to note that in these "standard" scores the numerator doesn't stray too far above the numerator. The sillier, more outlandish scores appear to be quite rare even compared to the scores where the numerator is less than the denominator. This suggests that, on the whole, _**WeRateDogs**_ regards the dogs whose pictures have been submitted to them as being worthy of praise and adulation and that the account is quite lenient in the scores it gives them.
